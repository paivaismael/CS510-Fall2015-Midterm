# Question 1

How does the plotted solution depend on your choice of time step size, and your choice of increment?  Do you see an improvement in precision from using the higher-order integration methods?

```python
a=Attractor()
a.evolve()
a.plotx()
a.ploty()
a.plotz()
a.plotxy()
a.plotyz()
a.plotzx()
a.plot3d()
```

```python
a=Attractor()
a.evolve(0.1,0.,0.,2)
a.plotx()
a.ploty()
a.plotz()
a.plotxy()
a.plotyz()
a.plotzx()
a.plot3d()
```

```python
a=Attractor()
a.evolve(0.1,0.,0.,1)
a.plotx()
a.ploty()
a.plotz()
a.plotxy()
a.plotyz()
a.plotzx()
a.plot3d()
```

Choosing a lower order, we have worse plots (i.e., plots with worse quality), because we are not using a good aproximation for the derivation of the function.

```python
a=Attractor(10.,28.,8./3.,0.,80.,5000)
a.evolve()
a.plotx()
a.ploty()
a.plotz()
a.plotxy()
a.plotyz()
a.plotzx()
a.plot3d()
```

```python
a=Attractor(10.,28.,8./3.,0.,80.,1000)
a.evolve()
a.plotx()
a.ploty()
a.plotz()
a.plotxy()
a.plotyz()
a.plotzx()
a.plot3d()
```

We can see that the time step size is very important. Reducing it we lose definition in the plots (or completly descaracterize it).

# Question 2

How does the solution depend upon the initial conditions [x0, y0, z0]?  How small a change can you make while still reproducing roughly the same dynamical curves?

```python
a=Attractor()
a.evolve(0.101,1.,3.)
a.plotx()
a.ploty()
a.plotz()
a.plotxy()
a.plotyz()
a.plotzx()
a.plot3d()
```

```python
a=Attractor()
a.evolve(5.,2.,15.)
a.plotx()
a.ploty()
a.plotz()
a.plotxy()
a.plotyz()
a.plotzx()
a.plot3d()
```

```python
a=Attractor()
a.evolve(0.,5.,6.)
a.plotx()
a.ploty()
a.plotz()
a.plotxy()
a.plotyz()
a.plotzx()
a.plot3d()
```

In the plots above we can see that the initial values are very important in the problem. They completely change the solution.

```python
a=Attractor()
a.evolve(0.1,0.001,0.0)
a.plotx()
a.ploty()
a.plotz()
a.plotxy()
a.plotyz()
a.plotzx()
a.plot3d()
```

```python
a=Attractor()
a.evolve(0.1,0.0,0.001)
a.plotx()
a.ploty()
a.plotz()
a.plotxy()
a.plotyz()
a.plotzx()
a.plot3d()
```

```python
a=Attractor()
a.evolve(0.1001,0.0,0.0)
a.plotx()
a.ploty()
a.plotz()
a.plotxy()
a.plotyz()
a.plotzx()
a.plot3d()
```

```python
a=Attractor()
a.evolve(0.101,0.001,0.001)
a.plotx()
a.ploty()
a.plotz()
a.plotxy()
a.plotyz()
a.plotzx()
a.plot3d()
```

In the plots above (mainly in the plotx, ploty, plotz) we can clearly see that even a small "perturbation" in the values of the inputs for s, p and b changes completely the solution of the problem.
